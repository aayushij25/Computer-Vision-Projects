### Select one of your photos and apply Average Blurring, Gaussian Blurring, Median Blurring, Bi-lateral Blurring, and show altogether. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def show_image(image, cmap = None, fig_size = (10, 10)):
    fig, ax = plt.subplots(figsize=fig_size)
    ax.imshow(image, cmap = cmap)
    ax.axis('off')
    plt.show()

image = cv2.imread('F:/Blurring/Picture_1.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

kernelSizes = [(51, 51)]
for i, (kX, kY) in enumerate(kernelSizes):
    # ----- Original image -----
    plt.figure(figsize=(20,10))
    plt.subplot(2,2,i+1),plt.imshow(gray, cmap = 'gray')
    plt.title('Original'), plt.xticks([]), plt.yticks([])
    # ----- Average Blurring -----
    # loop over the kernel sizes and apply an "average" blur to the image
    blurred = cv2.blur(gray, (kX, kY))
    plt.subplot(2,2,i+2),plt.imshow(blurred, cmap = 'gray')
    plt.title(f"Average ({kX}, {kY})"), plt.xticks([]), plt.yticks([])
    
    # ----- Original image -----
    plt.figure(figsize=(20,10))
    plt.subplot(2,2,i+3),plt.imshow(gray, cmap = 'gray')
    plt.title('Original'), plt.xticks([]), plt.yticks([])
    # ----- Gaussian Blurring -----
    # loop over the kernel sizes and apply an "Gaussian" blur to the image
    blurred1 = cv2.GaussianBlur(gray, (kX, kY), 0)
    plt.subplot(2,2,i+4),plt.imshow(blurred1, cmap = 'gray')
    plt.title(f"Gaussian ({kX}, {kY})"), plt.xticks([]), plt.yticks([])
    

# ----- Original image -----
plt.figure(figsize=(20,10))
plt.subplot(2,2,i+1),plt.imshow(gray, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
# ----- Median Blurring -----
kernel_sizes = [51]
for i, k in enumerate(kernel_sizes):
    # loop over the kernel sizes and apply an "median" blur to the image
    blurred2 = cv2.medianBlur(gray, k)
    plt.subplot(2,2,i+2),plt.imshow(blurred2, cmap = 'gray')
    plt.title(f"Median ({k})"), plt.xticks([]), plt.yticks([])

# ----- Original image -----
plt.figure(figsize=(20,10))
plt.subplot(2,2,i+1),plt.imshow(gray, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
# ----- Bi-lateral Blurring -----    
params = [(25, 75, 75)]
for i, (diameter, sigma_colour, sigma_space) in enumerate(params):
    # loop over the diameter, sigma color, and sigma space
    blurred3 = cv2.bilateralFilter(gray, diameter, sigma_colour, sigma_space)
    plt.subplot(2,2,i+2),plt.imshow(blurred3, cmap = 'gray')
    plt.title(f"Blurred d={diameter}, sc={sigma_colour}, ss={sigma_space}"), plt.xticks([]), plt.yticks([])
    
plt.show()
```
