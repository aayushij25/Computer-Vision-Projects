### Use a random number generator to draw random number of shapes in various sizes over the canvas. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import random

def show_image(image, cmap = None, fig_size = (10, 10)):
    fig, ax = plt.subplots(figsize=fig_size)
    ax.imshow(image, cmap = cmap)
    ax.axis('off')
    plt.show()
    

canvas = np.zeros((300, 300, 3), dtype="uint8")

#(purple,cyan,green,red,white,blue)
col1 = [(255,0,255),(0,255,255),(0,255,0),(255, 0, 0),(255, 255, 255),(0,0,255)]    
rancol = random.choice(col1)

angle = random.randint(8,90)
s1 = cv2.ellipse(canvas, (130, 130), (8,13), angle, 0 , 360, rancol, 2)

p1, p2, p3 = (119, 137), (129, 137), (random.randint(109,130), random.randint(155,158)) 
s2 = cv2.line(canvas, p1, p2, rancol, 2)
s3 = cv2.line(canvas, p2, p3, rancol, 2)
s4 = cv2.line(canvas, p1, p3, rancol, 2)

pts = np.array([[130,140], [140,140], [130,200], [120,200]], np.int32)
pts = pts.reshape((-1,1,2))
isClosed = True
s5 = cv2.polylines(canvas, [pts], isClosed, rancol, 2)

p4 = np.array([[118, 195], [128,183], [140,205], [132,220]], np.int32)
p4 = p4.reshape((-1,1,2))
isClosed = True
s6 = cv2.polylines(canvas, [p4], isClosed, rancol, 2)

s7 = cv2.ellipse(canvas, (167, 210), (24,36), 80, 0, 360, rancol, 2)

p5, p6, p7 = (208, 210), (random.randint(230,240), random.randint(225,240)), (190,220)
s8 = cv2.line(canvas, p5, p6, rancol, 2)
s9 = cv2.line(canvas, p6, p7, rancol, 2)
s10 = cv2.line(canvas, p5, p7, rancol, 2)

p8 = np.array([[167,231],[170,231],[171,258],[168,258]], np.int32)
p8 = p8.reshape((-1,1,2))
isClosed = True
s11 = cv2.polylines(canvas, [p8], isClosed, rancol, 2)

p9 = np.array([[168,231],[171,231],[181,258],[179,258]], np.int32)
p9 = p9.reshape((-1,1,2))
isClosed = True
s12 = cv2.polylines(canvas, [p9], isClosed, rancol, 2)

p10 = np.array([[179,260],[181,262],[random.randint(178,181),290],[random.randint(175,178),288]], np.int32)
p10 = p10.reshape((-1,1,2))
isClosed = True
s13 = cv2.polylines(canvas, [p10], isClosed, rancol, 2)

p11 = np.array([[168,260],[171, 258],[random.randint(153,170),290],[random.randint(150,167),288]], np.int32)
p11 = p11.reshape((-1,1,2))
isClosed = True
s14 = cv2.polylines(canvas, [p11], isClosed, rancol, 2)

for s15 in range(random.randint(50,100)):
    cv2.circle(canvas,(random.randint(0,300),random.randint(0,210)),1,(255, 0, 255),-1)


show_image(np.flip(canvas, axis=2))
```
